```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn import metrics
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.linear_model import LinearRegression, LogisticRegression
%matplotlib inline
```

```python
music_data = pd.read_csv ("mxmh_survey_results.csv")
music_data
```

```python
# What is the most common streaming service?
mxmh = music_data.groupby ("Primary streaming service")
mxmh = mxmh.sum()
mxmh = mxmh.sort_values (["Primary streaming service"], ascending = False)
mxmh.head (1)
```

```python
# What is the correlation between the following features: Hours per day, OCD,  and Anxiety
music_anxiety = music_data [[ "Hours per day", "OCD", "Anxiety"]]
music_anxiety
sns.heatmap(music_anxiety.corr(),annot = True, cmap = "magma" )
```

```python
# What is the average age of people that were interviewed?
average_age = music_data["Age"].mean()
average_age
```

```python
#Most popular genre?
mxmh = music_data.groupby ("Fav genre")
mxmh= mxmh.sum()
mxmh = mxmh.sort_values (["Fav genre"], ascending = False)
mxmh. head (1)
```
